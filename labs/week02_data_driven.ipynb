```python
import random

article = ["the", "a"]
noun = ["man", "ball", "woman", "table"]
verb = ["hit", "took", "saw", "liked"]

def make_noun_phrase():
    return [random.choice(article), random.choice(noun)]

def make_verb_phrase():
    return [random.choice(verb)] + make_noun_phrase()

def make_sentence():
    return make_noun_phrase() + make_verb_phrase()

def print_sentence():
    words = make_sentence()
    sentence = " ".join(words).capitalize() + "."
    print(sentence)

print_sentence()
```

### Poem generation

Using the following structure, generate a 5 line Haiku poem.

```python
import random

GRAMMAR = {
    "poem": [
        "noun_phrase",
        "verb_phrase",
        "preposition_phrase",
        "adjective_phrase",
        "noun_phrase"
    ],

    "sentence": ["noun_phrase verb_phrase"],

    "noun_phrase": ["article noun"],
    "verb_phrase": ["verb noun_phrase"],
    "preposition_phrase": ["preposition noun_phrase"],
    "adjective_phrase": ["intensifier adjective"],

    "preposition": ["after", "before", "on", "under", "inside", "outside"],
    "adjective": ["gentle", "curious", "Helpful", "Friendly", "Tall", "Short", "Clever", "Generous"],
    "intensifier": ["absolutely", "completely", "extremely", "highly", "really"],
    "article": ["the", "a"],
    "noun": ["man", "ball", "woman", "table"],
    "verb": ["hit", "took", "saw", "liked"]
}

def expand(x):
    if x not in GRAMMAR:
        return [x]
    
    output = random.choice(GRAMMAR[x])

    result = []
    for y in output.split():
        result += expand(y)  
    return result     

def generate_sentence():
    words = expand("sentence")
    return " ".join(words).capitalize() + "."

def generate_poem():
    lines = []

    for x in GRAMMAR["poem"]:
        lines.append(" ".join(expand(x)).capitalize())
    return "\n".join(lines)

print(generate_poem())
print(generate_sentence())

```
